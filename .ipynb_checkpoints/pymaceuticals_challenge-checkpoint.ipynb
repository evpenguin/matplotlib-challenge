```python
#Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

#open and read data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

#combine the data into a single DataFrame
mouse_study_df = pd.merge(study_results, mouse_metadata, on='Mouse ID', how='left')

#display dataframe head
mouse_study_df.head()
```

```python
#display the number of unique mouse id's
mouse_study_df['Mouse ID'].nunique()
```

```python
#locates all rows that are duplicates in mouse ID and timepoint and prints them
duplicate_mice = mouse_study_df.loc[mouse_study_df.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False), :]
duplicate_mice
```

```python
#gets list of indexes for mouse g989
mouse_index = mouse_study_df[(mouse_study_df['Mouse ID']=='g989')].index

#drops all rows for mouse g989
mouse_study_cleaned = mouse_study_df.drop(mouse_index)

#displays the head of the cleaned dataframe
mouse_study_cleaned.head()
```

```python
#display the number of unique mouse id's, to confirm we have successfully dropped the repeated mouse
mouse_study_cleaned['Mouse ID'].nunique()
```

```python
#groupby drug type
drug_group = mouse_study_cleaned.groupby(mouse_study_cleaned['Drug Regimen'])

#calculate summary stats
mean_tumor_volume = drug_group['Tumor Volume (mm3)'].mean()
median_tumor_volume = drug_group['Tumor Volume (mm3)'].median()
variance_tumor_volume = drug_group['Tumor Volume (mm3)'].var()
std_tumor_volume = drug_group['Tumor Volume (mm3)'].std()
sem_tumor_volume = drug_group['Tumor Volume (mm3)'].sem()

#merge summary stats into a DataFrame, renaming columns as we go
summary_stats = pd.concat([mean_tumor_volume.rename('Mean Tumor Volume'), median_tumor_volume.rename('Median Tumor Volume'), 
                           variance_tumor_volume.rename('Tumor Volume Variance'), std_tumor_volume.rename('Tumor Volume Std. Dev.'),
                           sem_tumor_volume.rename('Tumor Volume Std. Err.')], axis=1)

#display the summary stats dataframe
summary_stats
```

```python
#alternate way of calculating the summary stats
summary_stats_alt = drug_group.agg({'Tumor Volume (mm3)':['mean', 'median', 'var', 'std', 'sem']})
summary_stats_alt
```

```python
#group by mouse id
mouse_group = mouse_study_cleaned.groupby('Mouse ID')

#merge with original data to create a dataframe that contains the latest data for each mouse
final_timepoints_df = pd.merge(mouse_group.max()['Timepoint'], mouse_study_cleaned, on=['Mouse ID','Timepoint'], how='left')
```

```python
#list of treatments we want to explore further
poi_treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

#creates empty dataframe for use in the loop
final_tumor_volume = pd.DataFrame()

#loop through each drug we are interested in
for drug in poi_treatments:
    #find the final tumor volumes for each mouse on the drug
    tumor_volumes = final_timepoints_df.loc[final_timepoints_df['Drug Regimen'] == drug, 'Tumor Volume (mm3)']
    #add new column of those values, headed with drug name
    final_tumor_volume[drug] = tumor_volumes.values

#creates empty dataframe for use in the loop
poi_outliers = pd.DataFrame()

#loop through each drug we are interested in
for drug in poi_treatments:
    #calculate quartiles, iqr, upper and lower bounds
    quartiles = final_tumor_volume.quantile([0.25, 0.5, 0.75])
    Q1 = quartiles.at[0.25, drug]
    Q3 = quartiles.at[0.75, drug]
    iqr = Q3 - Q1
    upper_bound = Q3 + 1.5*iqr
    lower_bound = Q1 - 1.5*iqr
    #add new column containing any outliers, headed with drug name
    poi_outliers[drug] = final_tumor_volume.loc[(final_tumor_volume[drug] >= upper_bound) | (final_tumor_volume[drug] <= lower_bound), drug]

#print outliers data
poi_outliers
```

```python
#create boxplot of final tumor data
plt.boxplot(final_tumor_volume, flierprops={'marker': 'o', 'markerfacecolor': 'red', 'markersize': 12}, labels=poi_treatments)
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

```python
#create line graph of timepoint v tumor volume for mouse L509
plt.plot(mouse_study_cleaned.loc[mouse_study_cleaned['Mouse ID'] == 'l509', 'Timepoint'], mouse_study_cleaned.loc[mouse_study_cleaned['Mouse ID'] == 'l509', 'Tumor Volume (mm3)'])
plt.xlabel('Timepoint (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Capomulin treatment of mouse L509')
plt.show()
```

```python
#filter cleaned data for only mice on capomulin
capomulin_mice = mouse_study_cleaned.loc[mouse_study_cleaned['Drug Regimen'] == 'Capomulin', :]

#group the capomulin data by mouse
capo_group = capomulin_mice.groupby('Mouse ID')

#scatter plot of mouse weight v average tumor volume
plt.scatter(capo_group.mean()['Weight (g)'], capo_group.mean()['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

```python
#calculate correlation for weight and mean tumor volume
weightvtumor = st.pearsonr(capo_group.mean()['Weight (g)'], capo_group.mean()['Tumor Volume (mm3)'])

#print correlation value
print(f'The correlation between mouse weight and the average tumor volume is {round(weightvtumor[0], 2)}')

#caluclate linear regression variables
corr_slope, corr_int, corr_r, corr_p, corr_std_err = st.linregress(capo_group.mean()['Weight (g)'], capo_group.mean()['Tumor Volume (mm3)'])

#calculate regression line
corr_fit = (corr_slope * capo_group.mean()['Weight (g)']) + corr_int

#scatter plot
plt.scatter(capo_group.mean()['Weight (g)'], capo_group.mean()['Tumor Volume (mm3)'])

#add regression line
plt.plot(capo_group.mean()['Weight (g)'], corr_fit, color='red')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
